```python
allData = pd.read_csv("./data/0cleanDatasetWithQualifiersWith2025.csv")
update_stats_param = {
    "k_factor": None, 
    "base_k_factor": 43, 
    "max_k_factor": 62, 
    "div_number": 800, 
    "bonus_after_layoff": True
}
```

```python
allData_upto_2025 = allData

prev_stats_eval = createStats()

# Update up until end of 2024
for index, row in tqdm(allData_upto_2025.iterrows(), total=len(allData_upto_2025)):
    ########## UPDATE STATS ##########
    prev_stats_eval = updateStats(row, prev_stats_eval, **update_stats_param)
```

```python
def plot_player_elo_graph(player_id, player_name):
    plt.close('all')

    player = list(prev_stats_eval["elo_grad_players"][player_id])
    # player_grass = list(prev_stats_eval["elo_surface_players"]["Grass"][player_id])
    # player_clay = list(prev_stats_eval["elo_surface_players"]["Clay"][player_id])
    # player_hard = list(prev_stats_eval["elo_surface_players"]["Hard"][player_id])

    # Create a simple line plot
    fig = plt.figure(figsize=(8, 4))

    # plt.plot(player_grass, marker='.', linewidth=0.5, markersize=1, linestyle='-', color='green', label="Grass")
    # plt.plot(player_clay, marker='.', linewidth=0.5, markersize=1, linestyle='-', color='brown', label="Clay")
    # plt.plot(player_hard, marker='.', linewidth=0.5, markersize=1, linestyle='-', color='#ADD8E6', label="Hard")
    plt.plot(player, marker='.', linewidth=0.5, markersize=1, linestyle='-', color='black', label="Overall")


    plt.title(player_name+"'s Elo Rating Over Time")
    plt.xlabel("Match Number")
    plt.ylabel("Elo Rating")

    plt.legend(loc="lower right")

    plt.grid(True)
    plt.savefig("./images/"+"".join(player_name.split())+"_elo_ratingV2.png", dpi=500)

    plt.draw()
    plt.show()
```
